```python
#Función para eliminar dígitos
def eliminar_digitos( text ):
    pattern = r'[0-9]'
    text = re.sub( pattern, '', text )
    return text
```

# ***PAQUETE Nltk***
https://www.nltk.org/

```python
import nltk
nltk.download('popular')
```

```python
#Tokenizar el texto
from nltk.tokenize import word_tokenize
tokens= word_tokenize(texto)
tokens
```

```python
# Lista de stopwords: palabras vacías (artículos,preposiciones y conjunciones)
from nltk.corpus import stopwords
sw= stopwords.words('spanish') #Se configuran stopwords en español
sw
```

```python
#Función para limpieza de texto
import re
def limpiar_texto(texto):
  #1. Se convierte a minúscula
  texto=texto.lower()
  #2. Eliminar tildes
  replacements = (("á", "a"),("é", "e"),("í", "i"),("ó", "o"),("ú", "u"))
  for a, b in replacements:
      texto = texto.replace(a, b)
  # 3.Se eliminan dígitos y caracteres especiales
  pattern = r'[^a-z ]'
  texto = re.sub( pattern, '', texto )
  #4. Se tokeniza
  lista = word_tokenize(texto)
  #5. Se eliminan las stopwords
  clean_tokens = lista[:]
  for token in lista:
    if token in sw:
      clean_tokens.remove(token)
  return clean_tokens
```

```python
#Se llama la función de limpieza
tokens_clean=limpiar_texto(texto)
tokens_clean
```

```python
#Nube de palabras
texto_wc= ' '.join(tokens_clean)
wc = WordCloud(background_color="black").generate(texto_wc)
wc.to_file("nube1.png")
plt.axis("off")
plt.imshow(wc, interpolation='bilinear')
plt.show()
```

```python
# Calcular frecuencia de aparición de todas las palabras
from nltk.probability import FreqDist
fdist = FreqDist(tokens_clean)
fdist.plot(20)
```

```python
from nltk import bigrams

fdist = FreqDist(bigrams(tokens_clean))
fdist.most_common(10)
```

```python
fdist.plot(20)
```

```python
#Tri-gramas
from nltk.util import ngrams
fdist = FreqDist(ngrams(tokens_clean,3))
fdist.most_common(10)
```

```python
fdist.plot(10)
```

# **PAQUETE TextBlob**
https://textblob.readthedocs.io/en/dev/

```python
#Análisis de sentimientos
from textblob import TextBlob
```

```python
!python -m textblob.download_corpora
```

```python
#Análisis de sentimientos

comment='Im in bankruptcy and im going to lose my house.'
obj = TextBlob(comment)
sentimiento = obj.sentiment.polarity
if sentimiento > 0.0:
  print("positivo")
elif sentimiento < 0.0:
  print("negativo")
else:
    print("neutro")
```

```python
#Corrección de errores
b = TextBlob("I havv goood speling!")
print(b.correct())
```

```python
#Corrección de errores
b = TextBlob("I'm in bankruptcy and i'm going to lose my house.")
print(b.correct())
```
